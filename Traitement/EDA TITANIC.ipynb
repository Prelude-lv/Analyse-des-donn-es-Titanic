```python
from sklearn.impute import SimpleImputer
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('D:/AGTS/Bibliotheques and others/SEABORN/titanic.xls')
```

```python
data.head()
```

```python
## Vrification des valeurs manquantes  l'aide d'une carte thermique
import seaborn as sns
missing_values = data.isnull()
plt.figure(figsize=(10,7))
sns.heatmap(missing_values, cmap='viridis', cbar=True)
plt.show()
```

```python
## classification binaire avec la variable cible Survived
sns.countplot(x = 'survived', data=data)
```

```python
## Rpartition de la population dans chacune des classes
sns.countplot(x='pclass', data=data)
```

```python
## Rpartition et survivance des passagers !
sns.countplot(x='pclass', hue='survived', data=data)
```

```python
## Comparaison de comment le sexe a t affect
sns.catplot(x='sex', col='survived', kind='count', data=data)
```

```python
## verifions par classe
sns.catplot(x='pclass', col='survived', kind='count', data=data)
```

```python
## verifions par rapport au sexe
sns.catplot(x='sex', col='pclass', kind='count',data=data)
```

```python
## Analyse bivari
sns.catplot(x='pclass', y='survived', kind='point', hue='sex', data=data)
plt.show()
```
